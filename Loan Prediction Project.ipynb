```python
import pandas as pd
```

```python
df = pd.read_csv("Loan Prediction dataset.csv")
df.head(5)
```

```python
df = df.copy()
```

```python
df['Gender'] = df['Gender'].fillna('Male')
df['Married'] = df['Married'].fillna('Yes')
df['Dependents'] = df['Dependents'].fillna('0')
df['Self_Employed'] = df['Self_Employed'].fillna('No')
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(360)
df['Credit_History'] = df['Credit_History'].fillna(1)
```

```python
df.head(10)
```

```python
df['Total_Income'] = df['ApplicantIncome'] + df['CoapplicantIncome']
df['Loan_Income_Ratio'] = df['LoanAmount'] / df['Total_Income']
```

```python
def predict_loan(row):
    if row['Credit_History'] == 0:
        return 'Rejected'
    if row['Loan_Income_Ratio'] > 0.25:
        return 'Rejected'
    if row['Self_Employed'] == 'Yes' and row['LoanAmount'] > 300:
        return 'Rejected'
    return 'Approved'

df['Loan_Status_Predicted'] = df.apply(predict_loan, axis=1)
```

```python
# Export
df.to_excel("loan_predictions_analysis.xlsx", index=False)

print("Done! File saved: loan_predictions_analysis.xlsx")
```
